```python
import pymysql
import pandas as pd 
```

```python
db_settings = {
    'host': 'localhost',  # Change if necessary
    'port': 3306,         # Default MySQL port
    'user': 'root',       # Your MySQL username
    'password': 'root',   # Your MySQL password
    'database': 'stock_data'  # Replace with your database name
}
```

```python
connection = pymysql.connect(**db_settings)
```

```python
try:
    with connection.cursor() as cursor:
        # Write your SQL query
        sql_query = "SELECT * FROM stock_data_analyst order by datetime asc"  # Adjust as needed
        
        # Execute the query
        cursor.execute(sql_query)
        
        # Fetch all the results
        results = cursor.fetchall()
        
        # Print the results
        for row in results:
            print(row)
        column_names = [desc[0] for desc in cursor.description]
        
        # Create DataFrame with column names
        df = pd.DataFrame(list(results), columns=column_names)
        # Hiển thị các dòng đầu tiên
        print(df.head(30))
finally:
    # Close the connection
    connection.close()
```

```python
df.to_csv('stocks_data.csv', index=False)
```
